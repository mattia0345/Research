```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy.integrate import odeint

def dEdt(t, E, k1, k2, k3, r_a, r_c, a_tild_0, c_tild_0, V):
    a, b, c = E
    return [(k1*(b - a) + k2*(b**2 - a*b) - r_a*a + a_tild_0*r_a),
            (k1*(a - b) + k2*(a*b - b**2) + k3*(c - b)),
            (k3*(b - c) - r_c*c + c_tild_0*r_c)] 
a0 = 1; b0 = 2; c0 = 3
r_a = 1; r_c = 1; a_tild_0 = 2; c_tild_0 = 4

dt = 0.001

time = 1000000
time_array = np.arange(0, time*dt, dt)

V = 50
sol = odeint(dEdt, y0 = (a0, b0, c0), t = time_array, args = (1, 1, 1, r_a, r_c, a_tild_0, c_tild_0, V), tfirst=True) 
sol = sol.T 

fig, axes = plt.subplots(1, 3, sharey=True, figsize=(12,5))

axes[0].set_ylabel("Concentration")
axes[0].plot(time_array, sol[0], label='[A]', color='blue')
axes[1].plot(time_array, sol[1], label='[B]', color='red')
axes[2].plot(time_array, sol[2], label='[C]', color='green')
for ax in axes:
    ax.set_xlabel("Time")
    ax.set_xscale('log')
    ax.legend()

plt.minorticks_on()
plt.tight_layout()
plt.show()
```

# SOLIVNG CONCENTRATION ODES

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy.integrate import odeint

def dNdt(t, E, k1, k2, k3, r_a, r_c, a_tild_0, c_tild_0, V):
    a, b, c = E
    return [V * (k1*(b - a) + k2*(b**2 - a*b) - r_a*a + a_tild_0*r_a),
            V * (k1*(a - b) + k2*(a*b - b**2) + k3*(c - b)),
            V * (k3*(b - c) - r_c*c + c_tild_0*r_c)] 
a0 = 1; b0 = 2; c0 = 3
r_a = 1; r_c = 1; a_tild_0 = 2; c_tild_0 = 4

dt = 0.001

time = 1000000
time_array = np.arange(0, time*dt, dt)

V = 10
sol = odeint(dNdt, y0 = (a0, b0, c0), t = time_array, args = (1, 1, 1, r_a, r_c, a_tild_0, c_tild_0, V), tfirst=True) 
sol = sol.T 

fig, axes = plt.subplots(1, 3, sharey=True, figsize=(12,5))

axes[0].set_ylabel("Counts")
axes[0].plot(time_array, sol[0], label='[A]', color='blue')
axes[1].plot(time_array, sol[1], label='[B]', color='red')
axes[2].plot(time_array, sol[2], label='[C]', color='green')
for ax in axes:
    ax.set_xlabel("Time")
    ax.set_xscale('log')
    ax.legend()

plt.minorticks_on()
plt.tight_layout()
plt.show()
```
