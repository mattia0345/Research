```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()
G.add_node(1)
# G.add_nodes_from([2, 3])
G.add_edge(1, 3, weight = 2)
e = (2, 3); G.add_edge(*e)  # unpack edge tuple*
nx.draw(G, with_labels=True)
```
