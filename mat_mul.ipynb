```python
def padded_binary(i: int, n: int) -> str:
    return bin(i)[2:].zfill(n)

@staticmethod
def binary_string_to_array(string: str) -> np.ndarray:
    return np.array([int(i) for i in string], dtype=int)

def calculate_valid_transitions(self) -> Tuple[List[List[Any]], List[List[Any]]]:
    """
    Returns:
        valid_X_reactions: list of [state_i_str, state_j_str, i, j, "E"]
        valid_Y_reactions: list of [state_i_str, state_j_str, i, j, "F"]
    """
    all_states = [self.padded_binary(i, self.n) for i in range(self.num_states)]

    valid_difference_vectors = set()
    valid_X_reactions = []
    valid_Y_reactions = []

    for i in range(self.num_states):
        arr_i = self.binary_string_to_array(all_states[i])
        for j in range(self.num_states):
            if i == j:
                continue
            arr_j = self.binary_string_to_array(all_states[j])
            diff = arr_j - arr_i
            # if self.reaction_types == "distributive":
                
            hamming_weight = np.sum(np.abs(diff))

            if hamming_weight == 1:
                # +1 -> phosphorylation (E), -1 -> dephosphorylation (F)
                element = "E" if np.any(diff == 1) else "F"
                if element == "E":
                    if self.verbose:
                        print(f"{all_states[i]} --> {all_states[j]} (E), {i}, {j}")
                    valid_X_reactions.append([all_states[i], all_states[j], i, j, element])
                else:
                    if self.verbose:
                        print(f"{all_states[i]} --> {all_states[j]} (F), {i}, {j}")
                    valid_Y_reactions.append([all_states[i], all_states[j], i, j, element])
                valid_difference_vectors.add(tuple(diff))

    return valid_X_reactions, valid_Y_reactions
```

```python
import numpy as np

def matrix_normalize(matrix: np.ndarray) -> np.ndarray:
    """
    Normalizes the non-zero entries of a NumPy matrix so that the mean of 
    all non-zero entries equals 1.

    Zero entries are left unchanged.

    Args:
        matrix: A NumPy array (matrix) containing numerical data.

    Returns:
        A new NumPy array with the non-zero entries normalized, or the 
        original matrix if normalization is not possible (e.g., mean is zero).
    """

    # Create a copy to ensure the original matrix is not modified
    normalized = matrix.copy()

    # 1. Create a boolean mask for all non-zero elements
    non_zero_mask = normalized != 0

    # 2. Extract the non-zero elements
    non_zero_elements = normalized[non_zero_mask]

    # Handle the edge case where the matrix is all zeros or empty
    if non_zero_elements.size == 0:
        print("Warning: Matrix contains no non-zero elements. Returning original copy.")
        return normalized

    # 3. Calculate the mean of only the non-zero elements
    mean_non_zero = non_zero_elements.mean()

    # Handle the edge case where the mean of non-zero elements is 0
    if mean_non_zero == 0:
        print("Warning: Mean of non-zero elements is zero. Cannot normalize (division by zero). Returning original copy.")
        return normalized

    # 4. Normalize the non-zero elements by dividing them by the calculated mean
    normalized[non_zero_mask] = normalized[non_zero_mask] / mean_non_zero

    return normalized

# --- Example Usage ---

# 1. Create a sample matrix with some zeros
sample_matrix = np.array([
    [10.0, 20.0,  0.0],
    [30.0, 40.0, 50.0],
    [ 0.0, 60.0, 70.0]
])

print("--- Original Matrix ---")
print(sample_matrix)

# Non-zero entries: 10, 20, 30, 40, 50, 60, 70
# Sum of non-zero entries: 10 + 20 + 30 + 40 + 50 + 60 + 70 = 280
# Count of non-zero entries: 7
# Mean of non-zero entries: 280 / 7 = 40.0

# 2. Apply the normalization function
normalized_matrix = matrix_normalize(sample_matrix)

print("\n--- Normalized Matrix ---")
print(normalized_matrix)

# 3. Verify the result: Calculate the mean of the non-zero elements in the normalized matrix
mask_normalized = normalized_matrix != 0
non_zero_normalized = normalized_matrix[mask_normalized]

if non_zero_normalized.size > 0:
    new_mean = non_zero_normalized.mean()
    print(f"\nMean of non-zero elements in normalized matrix: {new_mean:.4f}")
    
    # Check if the new mean is approximately 1
    if np.isclose(new_mean, 1.0):
        print("Normalization successful: The new mean is approximately 1.")
    else:
        print("Normalization failed to produce a mean of 1.")

# Example with all zeros
print("\n--- Example with All Zeros ---")
zero_matrix = np.zeros((2, 2))
normalized_zero = matrix_normalize(zero_matrix)
print("Normalized All-Zero Matrix:")
print(normalized_zero)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# --- Gamma Distribution Parameters ---
# The shape parameter (alpha, often denoted as 'a' in scipy)
shape_param = 1
# The scale parameter (theta)
scale_param = 10 

# --- 1. Define the domain (x-values) ---
# We'll plot from 0 up to 20, using 100 points
x = np.linspace(0, 50, 1000)

# --- 2. Calculate the PDF (y-values) ---
# The scipy.stats.gamma.pdf function takes the x-values, shape ('a'), and scale
pdf_values = gamma.pdf(x, a=shape_param, scale=scale_param)

# --- 3. Plot the results ---
plt.figure(figsize=(8, 5))
plt.plot(x, pdf_values, label=f'Shape () = {shape_param}, Scale () = {scale_param}', color='teal')

# Add labels and title for clarity
plt.title(f'Probability Density Function (PDF) of the Gamma Distribution')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True, alpha=0.3)
plt.legend()

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

def plot_gamma_distribution(shape, scale, num_points=1000, num_samples=10000):
    """
    Plot the gamma distribution PDF and histogram of random samples.
    
    Parameters:
    -----------
    shape : float
        Shape parameter (k or alpha) of the gamma distribution
    scale : float
        Scale parameter (theta) of the gamma distribution
    num_points : int
        Number of points for plotting the PDF curve
    num_samples : int
        Number of random samples to generate for histogram
    """
    
    # Create figure with two subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    
    # Generate x values for PDF
    # Use mean  4*std as a reasonable range
    mean = shape * scale
    std = np.sqrt(shape) * scale
    x_min = max(0, mean - 4*std)
    x_max = mean + 4*std
    x = np.linspace(x_min, x_max, num_points)
    
    # Calculate PDF
    pdf = gamma.pdf(x, a=shape, scale=scale)
    
    # Plot 1: PDF curve
    ax1.plot(x, pdf, 'b-', linewidth=2, label='PDF')
    ax1.fill_between(x, pdf, alpha=0.3)
    ax1.axvline(mean, color='r', linestyle='--', linewidth=2, label=f'Mean = {mean:.4f}')
    ax1.axvline(mean - std, color='g', linestyle=':', linewidth=1.5, alpha=0.7)
    ax1.axvline(mean + std, color='g', linestyle=':', linewidth=1.5, alpha=0.7, 
                label=f'Std = {std:.4f}')
    ax1.set_xlabel('x', fontsize=12)
    ax1.set_ylabel('Probability Density', fontsize=12)
    ax1.set_title(f'Gamma Distribution PDF\nShape={shape}, Scale={scale}', fontsize=14, fontweight='bold')
    ax1.legend(fontsize=10)
    ax1.grid(True, alpha=0.3)
    
    # Plot 2: Histogram of samples
    samples = np.random.gamma(shape, scale, num_samples)
    ax2.hist(samples, bins=50, density=True, alpha=0.7, color='skyblue', edgecolor='black')
    ax2.plot(x, pdf, 'r-', linewidth=2, label='Theoretical PDF')
    ax2.axvline(np.mean(samples), color='orange', linestyle='--', linewidth=2, 
                label=f'Sample Mean = {np.mean(samples):.4f}')
    ax2.set_xlabel('x', fontsize=12)
    ax2.set_ylabel('Probability Density', fontsize=12)
    ax2.set_title(f'Histogram of {num_samples} Random Samples', fontsize=14, fontweight='bold')
    ax2.legend(fontsize=10)
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # Print statistics
    print(f"\n{'='*50}")
    print(f"Gamma Distribution Statistics")
    print(f"{'='*50}")
    print(f"Parameters:")
    print(f"  Shape (k):     {shape}")
    print(f"  Scale ():     {scale}")
    print(f"\nTheoretical:")
    print(f"  Mean:          {mean:.6f}")
    print(f"  Variance:      {shape * scale**2:.6f}")
    print(f"  Std Dev:       {std:.6f}")
    print(f"  Mode:          {max(0, (shape-1)*scale):.6f}" + 
          (" (undefined for shape < 1)" if shape < 1 else ""))
    print(f"\nSample Statistics ({num_samples} samples):")
    print(f"  Mean:          {np.mean(samples):.6f}")
    print(f"  Variance:      {np.var(samples):.6f}")
    print(f"  Std Dev:       {np.std(samples):.6f}")
    print(f"  Min:           {np.min(samples):.6f}")
    print(f"  Max:           {np.max(samples):.6f}")
    print(f"  Median:        {np.median(samples):.6f}")
    print(f"{'='*50}\n")
    
    return samples


def plot_multiple_gamma_distributions(param_list, num_points=1000):
    """
    Plot multiple gamma distributions on the same axes for comparison.
    
    Parameters:
    -----------
    param_list : list of tuples
        List of (shape, scale, label) tuples
    num_points : int
        Number of points for plotting each PDF curve
    """
    
    plt.figure(figsize=(12, 6))
    
    colors = plt.cm.Set2(np.linspace(0, 1, len(param_list)))
    
    for i, (shape, scale, label) in enumerate(param_list):
        mean = shape * scale
        std = np.sqrt(shape) * scale
        x_min = max(0, mean - 4*std)
        x_max = mean + 4*std
        x = np.linspace(x_min, x_max, num_points)
        
        pdf = gamma.pdf(x, a=shape, scale=scale)
        plt.plot(x, pdf, linewidth=2.5, label=label, color=colors[i])
        plt.fill_between(x, pdf, alpha=0.15, color=colors[i])
    
    plt.xlabel('x', fontsize=12)
    plt.ylabel('Probability Density', fontsize=12)
    plt.title('Comparison of Gamma Distributions', fontsize=14, fontweight='bold')
    plt.legend(fontsize=10)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()


# Example usage with your parameters from the code
if __name__ == "__main__":
    
    # Your old shape parameters
    print("OLD SHAPE PARAMETERS:")
    old_shape = 1
    old_scale = 3
    samples_old = plot_gamma_distribution(old_shape, old_scale)
    
    # Your new shape parameters
    # print("\nNEW SHAPE PARAMETERS:")
    # new_shape = 5e-1
    # new_scale = 1 / new_shape
    # samples_new = plot_gamma_distribution(new_shape, new_scale)
    
    # Compare multiple distributions
    # print("\nCOMPARISON PLOT:")
    param_list = [
        (0.123, 4.46e6, 'Old: shape=0.123, scale=4.46e6'),
        (1e6, 1e6, 'New: shape=1e6, scale=1e6'),
        (1, 1, 'Standard: shape=1, scale=1'),
        (2, 2, 'Moderate: shape=2, scale=2')
    ]
    plot_multiple_gamma_distributions(param_list)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

def plot_gamma_distribution(shape, scale, num_points=1000, num_samples=10000, normalize=False):
    """
    Plot the gamma distribution PDF and histogram of random samples.
    
    Parameters:
    -----------
    shape : float
        Shape parameter (k or alpha) of the gamma distribution
    scale : float
        Scale parameter (theta) of the gamma distribution
    num_points : int
        Number of points for plotting the PDF curve
    num_samples : int
        Number of random samples to generate for histogram
    normalize : bool
        If True, normalize samples to have mean of 1 after drawing
    """
    
    # Create figure with two subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    
    # Generate samples first
    samples = np.random.gamma(shape, scale, num_samples)
    
    # Normalize if requested
    if normalize:
        original_mean = np.mean(samples)
        samples = samples / original_mean
        print(f"Normalized samples: divided by original mean = {original_mean:.6e}")
    
    # Generate x values for PDF
    # Use mean  4*std as a reasonable range
    mean = shape * scale
    std = np.sqrt(shape) * scale
    
    if normalize:
        # Adjust theoretical mean and std for normalized case
        mean = 1.0  # By definition after normalization
        # Std scales with the mean
        cv = std / (shape * scale)  # coefficient of variation
        std = cv * mean
    
    x_min = max(0, mean - 4*std)
    x_max = mean + 4*std
    x = np.linspace(x_min, x_max, num_points)
    
    # Calculate PDF
    if normalize:
        # For normalized case, we don't have a simple closed form
        # Just show the histogram
        pdf = None
    else:
        pdf = gamma.pdf(x, a=shape, scale=scale)
    
    # Plot 1: PDF curve (or just samples if normalized)
    if normalize:
        ax1.hist(samples, bins=50, density=True, alpha=0.7, color='skyblue', edgecolor='black')
        ax1.axvline(np.mean(samples), color='r', linestyle='--', linewidth=2, 
                    label=f'Sample Mean = {np.mean(samples):.4f}')
        ax1.axvline(1.0, color='orange', linestyle=':', linewidth=2, 
                    label='Target Mean = 1.0')
        ax1.set_title(f'Normalized Distribution\n(Original: Shape={shape}, Scale={scale})', 
                     fontsize=14, fontweight='bold')
    else:
        ax1.plot(x, pdf, 'b-', linewidth=2, label='PDF')
        ax1.fill_between(x, pdf, alpha=0.3)
        ax1.axvline(mean, color='r', linestyle='--', linewidth=2, label=f'Mean = {mean:.4f}')
        ax1.axvline(mean - std, color='g', linestyle=':', linewidth=1.5, alpha=0.7)
        ax1.axvline(mean + std, color='g', linestyle=':', linewidth=1.5, alpha=0.7, 
                    label=f'Std = {std:.4f}')
        ax1.set_title(f'Gamma Distribution PDF\nShape={shape}, Scale={scale}', 
                     fontsize=14, fontweight='bold')
    
    ax1.set_xlabel('x', fontsize=12)
    ax1.set_ylabel('Probability Density', fontsize=12)
    ax1.legend(fontsize=10)
    ax1.grid(True, alpha=0.3)
    
    # Plot 2: Histogram of samples
    ax2.hist(samples, bins=50, density=True, alpha=0.7, color='skyblue', edgecolor='black')
    if not normalize and pdf is not None:
        ax2.plot(x, pdf, 'r-', linewidth=2, label='Theoretical PDF')
    ax2.axvline(np.mean(samples), color='orange', linestyle='--', linewidth=2, 
                label=f'Sample Mean = {np.mean(samples):.4f}')
    ax2.set_xlabel('x', fontsize=12)
    ax2.set_ylabel('Probability Density', fontsize=12)
    title_suffix = " (Normalized)" if normalize else ""
    ax2.set_title(f'Histogram of {num_samples} Random Samples{title_suffix}', 
                 fontsize=14, fontweight='bold')
    ax2.legend(fontsize=10)
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # Print statistics
    print(f"\n{'='*50}")
    print(f"Gamma Distribution Statistics")
    if normalize:
        print(f"(NORMALIZED TO MEAN = 1)")
    print(f"{'='*50}")
    print(f"Original Parameters:")
    print(f"  Shape (k):     {shape}")
    print(f"  Scale ():     {scale}")
    
    if not normalize:
        print(f"\nTheoretical:")
        print(f"  Mean:          {shape * scale:.6f}")
        print(f"  Variance:      {shape * scale**2:.6f}")
        print(f"  Std Dev:       {np.sqrt(shape) * scale:.6f}")
        print(f"  Mode:          {max(0, (shape-1)*scale):.6f}" + 
              (" (undefined for shape < 1)" if shape < 1 else ""))
    
    print(f"\nSample Statistics ({num_samples} samples):")
    print(f"  Mean:          {np.mean(samples):.6f}")
    print(f"  Variance:      {np.var(samples):.6f}")
    print(f"  Std Dev:       {np.std(samples):.6f}")
    print(f"  Min:           {np.min(samples):.6f}")
    print(f"  Max:           {np.max(samples):.6f}")
    print(f"  Median:        {np.median(samples):.6f}")
    print(f"  Coef. of Var:  {np.std(samples)/np.mean(samples):.6f}")
    print(f"{'='*50}\n")
    
    return samples


def plot_multiple_gamma_distributions(param_list, num_points=1000):
    """
    Plot multiple gamma distributions on the same axes for comparison.
    
    Parameters:
    -----------
    param_list : list of tuples
        List of (shape, scale, label) tuples
    num_points : int
        Number of points for plotting each PDF curve
    """
    
    plt.figure(figsize=(12, 6))
    
    colors = plt.cm.Set2(np.linspace(0, 1, len(param_list)))
    
    for i, (shape, scale, label) in enumerate(param_list):
        mean = shape * scale
        std = np.sqrt(shape) * scale
        x_min = max(0, mean - 4*std)
        x_max = mean + 4*std
        x = np.linspace(x_min, x_max, num_points)
        
        pdf = gamma.pdf(x, a=shape, scale=scale)
        plt.plot(x, pdf, linewidth=2.5, label=label, color=colors[i])
        plt.fill_between(x, pdf, alpha=0.15, color=colors[i])
    
    plt.xlabel('x', fontsize=12)
    plt.ylabel('Probability Density', fontsize=12)
    plt.title('Comparison of Gamma Distributions', fontsize=14, fontweight='bold')
    plt.legend(fontsize=10)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()


# Example usage with your parameters from the code
if __name__ == "__main__":
    
    # Your old shape parameters - ORIGINAL
    print("="*60)
    print("OLD SHAPE PARAMETERS - ORIGINAL:")
    print("="*60)
    old_shape = 1
    old_scale = 3
    samples_old = plot_gamma_distribution(old_shape, old_scale, num_samples=10000)
    
    # Your old shape parameters - NORMALIZED
    print("\n" + "="*60)
    print("OLD SHAPE PARAMETERS - NORMALIZED TO MEAN = 1:")
    print("="*60)
    samples_old_normalized = plot_gamma_distribution(old_shape, old_scale, num_samples=10000, normalize=True)
    
    # Your new shape parameters
    print("\n" + "="*60)
    print("NEW SHAPE PARAMETERS:")
    print("="*60)
    new_shape = 2
    new_scale = 1 / new_shape
    samples_new = plot_gamma_distribution(new_shape, new_scale)
    
    # Direct comparison: normalized old vs new distributions
    print("\n" + "="*60)
    print("COMPARISON: Normalized Old vs New Parameters")
    print("="*60)
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    
    # Normalized old parameters
    ax1.hist(samples_old_normalized, bins=50, density=True, alpha=0.7, 
             color='coral', edgecolor='black', label='Normalized Old')
    ax1.axvline(np.mean(samples_old_normalized), color='red', linestyle='--', 
                linewidth=2, label=f'Mean = {np.mean(samples_old_normalized):.4f}')
    ax1.set_xlabel('x', fontsize=12)
    ax1.set_ylabel('Probability Density', fontsize=12)
    ax1.set_title('Normalized Old Parameters\n(shape=0.123, scale=4.46e6)', 
                  fontsize=14, fontweight='bold')
    ax1.legend(fontsize=10)
    ax1.grid(True, alpha=0.3)
    ax1.set_xlim([0, max(5, np.percentile(samples_old_normalized, 99))])
    
    # New parameters (normalized for fair comparison)
    samples_new_normalized = samples_new / np.mean(samples_new)
    ax2.hist(samples_new_normalized, bins=50, density=True, alpha=0.7, 
             color='skyblue', edgecolor='black', label='Normalized New')
    ax2.axvline(np.mean(samples_new_normalized), color='blue', linestyle='--', 
                linewidth=2, label=f'Mean = {np.mean(samples_new_normalized):.4f}')
    ax2.set_xlabel('x', fontsize=12)
    ax2.set_ylabel('Probability Density', fontsize=12)
    ax2.set_title('Normalized New Parameters\n(shape=1e6, scale=1e6)', 
                  fontsize=14, fontweight='bold')
    ax2.legend(fontsize=10)
    ax2.grid(True, alpha=0.3)
    ax2.set_xlim([0, max(5, np.percentile(samples_new_normalized, 99))])
    
    plt.tight_layout()
    plt.show()
    
    print("\nKey Observations:")
    print(f"  Old (normalized) CV:  {np.std(samples_old_normalized)/np.mean(samples_old_normalized):.4f}")
    print(f"  New (normalized) CV:  {np.std(samples_new_normalized)/np.mean(samples_new_normalized):.4f}")
    print(f"  Old 99th percentile:  {np.percentile(samples_old_normalized, 99):.4f}")
    print(f"  New 99th percentile:  {np.percentile(samples_new_normalized, 99):.4f}")
```

```python
import numpy as np

def matrix_normalize(matrix: np.ndarray) -> np.ndarray:
    normalized = matrix.copy()
    non_zero_mask = normalized != 0
    non_zero_elements = normalized[non_zero_mask]
    mean_non_zero = non_zero_elements.mean()
    normalized[non_zero_mask] = normalized[non_zero_mask] / mean_non_zero
    return normalized

# --- Example Usage ---

# 1. Create a sample matrix with some zeros
sample_matrix = np.array([
    [0.0, 1.0,  2.0, 0],
    [0.0, 0.0, 0.0, 3],
    [ 0.0, 0.0, 0.0, 4],
    [0, 0, 0, 0]
])

print("--- Original Matrix ---")
print(sample_matrix)

# Non-zero entries: 10, 20, 30, 40, 50, 60, 70
# Sum of non-zero entries: 10 + 20 + 30 + 40 + 50 + 60 + 70 = 280
# Count of non-zero entries: 7
# Mean of non-zero entries: 280 / 7 = 40.0

# 2. Apply the normalization function
normalized_matrix = matrix_normalize(sample_matrix)

print("\n--- Normalized Matrix ---")
print(normalized_matrix)

# 3. Verify the result: Calculate the mean of the non-zero elements in the normalized matrix
mask_normalized = normalized_matrix != 0
non_zero_normalized = normalized_matrix[mask_normalized]

if non_zero_normalized.size > 0:
    new_mean = non_zero_normalized.mean()
    print(f"\nMean of non-zero elements in normalized matrix: {new_mean:.4f}")
    
    # Check if the new mean is approximately 1
    if np.isclose(new_mean, 1.0):
        print("Normalization successful: The new mean is approximately 1.")
    else:
        print("Normalization failed to produce a mean of 1.")

# Example with all zeros
print("\n--- Example with All Zeros ---")
zero_matrix = np.zeros((2, 2))
normalized_zero = matrix_normalize(zero_matrix)
print("Normalized All-Zero Matrix:")
print(normalized_zero)
```

```python
import numpy as np
from scipy.optimize import root

def f(vec):
    x, y = vec
    return np.array([x**2 + y**2 - 1, x - y])

def deflated_f(vec, roots, alpha=1e-2, p=2):
    M = 1.0
    for r in roots:
        M *= 1.0 / (np.linalg.norm(vec - r)**p + alpha)
    return M * f(vec)

roots_found = []
x0_list = [np.random.randn(2) for _ in range(10)]  # random initial guesses

for x0 in x0_list:
    sol = root(lambda v: deflated_f(v, roots_found), x0)
    if sol.success:
        candidate = sol.x
        # check distinctness
        if all(np.linalg.norm(candidate - r) > 1e-6 for r in roots_found):
            roots_found.append(candidate)

print("Distinct roots found:")
for r in roots_found:
    print(r)
```
