```python
from julia import DifferentialEquations as de

def f(du, u, p, t):
    A, B, C = u
    k1, k2, k3, r_a, r_c, A_tild, C_tild = p
    du[0] = k1*(B - A) + k2*(0.5*B**2 - A*B) - r_a*A + r_a*A_tild
    du[1] = k1*(A - B) + k2*(A*B - 0.5*B**2) + k3*C - k3*B
    du[2] = k3*B - k3*C - r_c*C + r_c*C_tild

def g(du,u,p,t):    
    A, B, C = u
    k1, k2, k3, r_a, r_c, A_tild, C_tild = p
    du = np.array([[-np.sqrt(k1*u[0]), np.sqrt(k1*u[1]), 0],
                   ])
    # du[0] = np.sqrt(k1*u[1]) - np.sqrt(k1*u[0]) + 0.5*np.sqrt(k2*u[1]**2) - np.sqrt(k2*u[0]*u[1]) - np.sqrt(r_a*u[0]) + np.sqrt(r_a*A_tild)
    # du[1] = np.sqrt(k1*u[0]) - np.sqrt(k1*u[1]) + np.sqrt(k2*u[0]*u[1]) - 0.5*np.sqrt(k2*u[1]**2) + np.sqrt(k3*u[2]) - np.sqrt(k3*u[1])
    # du[2] = np.sqrt(k3*u[1]) - np.sqrt(k3*u[2]) - np.sqrt(r_c*u[2]) + np.sqrt(r_c*C_tild)

u0 = [1.0,2.0,3.0]
k1 = 1; k2 = 1; k3 = 1; r_a = 1; r_c = 1; a_tild = 2; c_tild = 4

tspan = (0., 100.)
p = (k1, k2, k3, r_a, r_c, a_tild, c_tild) 
prob = de.SDEProblem(f, g, u0, tspan, p)
sol = de.solve(prob)
print(sol)
# Now let's draw a phase plot
t = np.array(sol.t, dtype=float)
u = np.array([ui[0] for ui in sol.u], dtype=float)  # extract component 0
plt.plot(t, u)
# plt.xlabel("t"); plt.ylabel("u(t)"); plt.title("SDE: du = 1.01 u dt + 0.87 u dW")
plt.show()
```
